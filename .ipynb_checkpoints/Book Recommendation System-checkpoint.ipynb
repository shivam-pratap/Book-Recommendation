```python
rating_user.drop(['imageUrlS'],axis=1)
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(rating_user.loc[:, rating_user.columns != 'bookRating'], rating_user['bookRating'], stratify=rating_user['bookRating'], random_state=66)
```

```python
len(X_train)==len(y_train)#to check ambiguity in train test split
```

```python
X_train.columns
```

```python
from sklearn.linear_model import LogisticRegression

logreg = LogisticRegression().fit(X_train, y_train)
print("Training Socre:{:.3f}".format(logreg.score(X_train, y_train)))
print("Testing Socre:{:.3f}".format(logreg.score(X_test, y_test)))
```
